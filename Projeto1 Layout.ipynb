```python
dados.V02323 = dados.V02323.astype('category')
dados.V02324= dados.V02324.astype('category')
dados.V02325= dados.V02325.astype('category')


dados.V4621  = dados.V4621 .astype('float')

dados.V02323.cat.categories = (['Sim', 'Nao'])
dados.V02324.cat.categories = (['Sim', 'Nao'])
dados.V02325.cat.categories = (['Sim', 'Nao'])
```

```python
# Tabelas univariada

print('Frequências absolutas por Acessa internet do microcomputador:')
ut1 = dados.V02323.value_counts().reindex(['Sim', 'Nao'])
print(ut1,'\n')

print('Frequências absolutas por Para acessar à Internet utiliza-se telefone celular:')
ut1 = dados.V02324.value_counts().reindex(['Sim', 'Nao'])
print(ut1,'\n')

print('Frequências absolutas por Para acessar à Internet utiliza-se tablet:')
ut2 = dados.V02325.value_counts().reindex(['Sim', 'Nao'])
print(ut2,'\n')
```

```python
# Tabelas univariada
print('Frequências relativas por Acessa internet do microcomputador:')
ut1 = (dados.V02323.value_counts(normalize=True)*100).round(decimals=1).reindex(['Sim', 'Nao'])
print(ut1,'\n')

print('Frequências relativas por Para acessar à Internet utiliza-se telefone celular:')
ut2 = (dados.V02324.value_counts(normalize=True)*100).round(decimals=1).reindex(['Sim', 'Nao'])
print(ut2,'\n')

print('Frequências relativas por Para acessar à Internet utiliza-se tablet:')
ut3 = (dados.V02325.value_counts(normalize=True)*100).round(decimals=1).reindex(['Sim', 'Nao'])
print(ut3,'\n')
```

```python
plot = ut1.plot(kind='bar',title='Acessa internet do microcomputador',figsize=(6, 6),color=('b','g'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Acessa internet do microcomputador')
```

```python
plot = ut2.plot(kind='bar',title='Acessa internet do celular',figsize=(6, 6),color=('b','g'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Acessa internet do microcomputador')
```

```python
plot = ut3.plot(kind='bar',title='Acessa internet do tablet',figsize=(6, 6),color=('b','g'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Acessa internet do microcomputador')
```

```python
# Gráfico de colunas com frequência absoluta dos salarios
from numpy import arange
faixa = arange(800,7000,800)
plot = dados.V4621.plot.hist(bins=faixa,title='Grafico Salarios', figsize=(6, 6), alpha=0.5)
plt.ylabel('Frequência absoluta')
plt.xlabel('renda (reais)')
```
