###  Função com a finalidade de retornar diferentes faixas de renda per capita para serem plotadas nos decorrentes graficos. Nesta análise o foco foi em rendas de até mil reais por moradores, devido a renda apartir desse ponto não impedir a posse, ou até mesmo o acesso dessas pessoas ao respectivo meio.

```python

def faixasph(df):
    x=df['V4621']
    if x<100: return 1
    elif x<200: return 2
    elif x<300: return 3
    elif x<400: return 4
    elif x<500: return 5
    elif x<600: return 6
    elif x<700: return 7
    elif x<800: return 8
    elif x<900: return 9
    elif x<1000: return 10
    return 11
faixas=dados.apply(faixasph,axis=1)
dados['faixas']=faixas
dados.faixas=dados.faixas.astype('category')
dados.faixas.cat.categories=(['100','200','300','400','500','600','700','800','900','1000','+1000'])
```

```python
faixas=dados13.apply(faixasph,axis=1)
dados13['faixas']=faixas
dados13.faixas=dados.faixas.astype('category')
dados13.faixas.cat.categories=(['100','200','300','400','500','600','700','800','900','1000','+1000'])
```

### Nos dois seguintes gráficos, fica explícito que o aumento de renda influência positivamente no acesso à internet por meio de microcomputadores em 2015. Porém, em 2013 o computador se mostrou um eletrônico mais acessível para as camadas mais baixas, as quais chegaram a se igualar as camadas altas
### Sendo meuito peculiar a faixa de salários correspondente à oitocentos reais. A qual provavelmente se refere à uma faixa pouco pesquisada durante a formação da PNAD de 2015

```python
# Posse de computador
dados['Posse']=dados.V0231
ph=pd.crosstab(dados.faixas,dados.Posse,normalize='index')
ph.plot(kind='bar',stacked='True',legend=['Sim','Não'],title='Posse de computador x Renda (centena)-2015',figsize=(7,7))
plt.ylabel('Freq em %')
plt.xlabel('Renda em centena')


plt.plot()

# Posse de computador
dados13['Posse']=dados13.V0231
ph=pd.crosstab(dados13.faixas,dados13.Posse,normalize='index')
ph.plot(kind='bar',stacked='True',legend=['Sim','Não'],title='Posse de computador x Renda (centena)-2013',figsize=(7,7))
plt.ylabel('Freq em %')
plt.xlabel('Renda em centena')



plt.plot()
```
